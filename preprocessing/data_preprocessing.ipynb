To get around the issue of the sentiment being heavily swayed toward the opinion of a few tweets I normalised the sentiment with a function based off of the number of samples for each month:

$$-1 \cdot(\frac{1}{0.3\cdot x + 1} )+1$$

where x is the number of samples

Also for values that couldn't be interpolated I set them to the mean of the normalised values + a small amount of noise (1 std dev)

```python

def normalise(df1, df2):
    for index, row in df1.iterrows():
#         print(row)
        year = int(row['year'])
        month = int(row['month'])
        
        monthyear = str(year) + "-" + str(month)
#         print(monthyear)
        x = len(df2.loc[df2['monthyear'] == monthyear])
#         print(x)
        norm_factor = -(1/(0.3*x+1))+1
#         print(row['compound'])
#         print(row['compound'] * norm_factor)
        
        df1.at[index, 'compound'] = row['compound'] * norm_factor
    return df1

# dates = pd.to_datetime(sentiment_df['index'])
# dates = sentiment_df['index'].to_timestamp()
dates = pd.to_datetime(sentiment_df['index'].map(str))
dates
plt.plot(dates, sentiment_df['compound'], label="Pre-normalisation")
plt.xlabel("Date")
plt.ylabel("Average Sentiment")
plt.title("Average Sentiment VS Date")
plt.legend()
sentiment_normalised_df = normalise(sentiment_df, tweets_df)
plt.plot(dates, sentiment_normalised_df['compound'], label="Post-normalisation")
plt.legend()
```
